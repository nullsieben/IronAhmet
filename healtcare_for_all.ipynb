## standardizing header names

```python
def lower_case_column_names(file):
    file.columns=[i.lower() for i in file.columns]
    return file
```

## concat

```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'female',"Femal"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
hc['gender'] = list(map(clean_gender, hc['gender'])) 
```

```python
hc.gender.unique()
```

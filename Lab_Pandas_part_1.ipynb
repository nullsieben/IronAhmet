Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
## Your Code here
cars_df["CO2 Emission Grams/Mile"]=cars_df["CO2 Emission Grams/Mile"].apply(lambda x: x*1.60934)                        
```

```python
cars_df.rename(columns={"CO2 Emission Grams/Mile":"CO2 Emission Grams/Km"}, inplace=True)    
```

```python
cars_df.columns
```

```python
cars_df["City MPG"]=cars_df["City MPG"].apply(lambda x: (x*1.60934)/3.78541)  
```

```python
cars_df["Highway MPG"]=cars_df["Highway MPG"].apply(lambda x: (x*1.60934)/3.78541)  
cars_df["Combined MPG"]=cars_df["Combined MPG"].apply(lambda x: (x*1.60934)/3.78541)
```

```python
cars_df
```

```python
cars_df.rename(columns={"City MPG":"City km/l", "Highway MPG":"Highway km/l", "Combined MPG":"Combined km/l" }, inplace=True)
```
